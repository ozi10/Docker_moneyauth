```python
from sklearn.preprocessing import StandardScaler

Sc = StandardScaler()
Sc.fit(df)
```

```python
scaled_data = Sc.transform(df)
```

```python
df1 = pd.DataFrame(scaled_data)
df1.columns = ['variance','skewness','curtosis','entropy','class']
```

```python
df1.tail()
```

```python
sns.pairplot(df1,hue='class')
```

```python
X = df[['variance','skewness','curtosis','entropy']]
y = df['class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=20)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score,precision_recall_fscore_support as score
```

```python
%%time


Rf = RandomForestClassifier()
param = {
    'n_estimators' : [10,50,100],
    'max_depth' : [50,100,150]
}

Gs = GridSearchCV(Rf,param,cv=5,n_jobs=-1)
Gs_model = Gs.fit(X_train,y_train)
pd.DataFrame(Gs_model.cv_results_).sort_values('mean_test_score',ascending=False)[0:5]
```

```python
y_pred = Gs.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
import pickle
pickle_out = open("classifier.pkl",'wb')
pickle.dump(Gs,pickle_out)
pickle_out.close()
```
